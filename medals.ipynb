```python
import pandas as pd
import os
```

```python
medals_df = pd.read_csv('data/medals.csv')
```

```python
medals_df.head()
```

```python
print(len(medals_df))
```

```python
clean_medals_df = medals_df[["athlete_name",
                             "athlete_sex",
                             "medal_type",
                             "country",
                             "country_code",
                             "event",
                             "discipline",                                                          
                             "athlete_link"]]
```

```python
clean_medals_df.head()
```

```python
clean_medals_rename = clean_medals_df.rename(columns={"athlete_name":"name",
                                                      "athlete_sex":"gender",
                                                      "athlete_link":"url"})
```

```python
clean_medals_rename["medal_type"] = clean_medals_rename["medal_type"].str.replace(' Medal','')
clean_medals_rename["gender"] = clean_medals_rename["gender"].str.replace('M','Male').str.replace('W','Female').str.replace('X','Mixed')
clean_medals_rename["event"] = clean_medals_rename["event"].str.replace(' &Eacute;p&eacute;e', '')
```

```python
clean_medals_rename.head()
```

```python
print(len(clean_medals_rename))
```

```python
clean_medals_rename.to_csv('clean_data/clean_medals.csv',
                           encoding='utf-8', index=False, header=True)
```
